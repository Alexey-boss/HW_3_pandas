```python
import pandas as pd
```

```python
#Задание № 1

df_ratings = pd.read_csv(r'ratings.csv')
df_ratings.head()
```

```python
df_movies = pd.read_csv(r'movies.csv')
df_movies.head()
```

```python
df_all = df_ratings.merge(df_movies, left_on="movieId",right_on='movieId', how='left')
df_all.head()
```

```python
def rating(df):
    if df['rating'] <= 2:
        return "низкий рейтинг"
    elif df['rating'] <= 4:
        return "средний рейтинг"
    elif df['rating']==4.5 or df['rating'] == 5:
        return "высокий рейтинг"    
df_all['Class'] = df_all.apply(rating,axis=1)
```

```python
df_keywords = pd.read_csv(r'keywords.csv')
```

```python
# Задание № 2

def geo_class(df):
    geo_data = {
            'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
             }
    
    for i,k in geo_data.items():
        if  len(list(set(k) & set(df['keyword'].split(' '))))!=0:
            result = i
            break
        else:   
            result= "undefined"
    return result
              
df_keywords['region'] = df_keywords.apply(geo_class, axis=1)
```
